```python
import pandas as pd
import numpy as np
```

```python
# Upload data
data=pd.read_csv('movie_data.csv')
data.columns
```

```python
# Only keep those features that we required

data=data[['director_name','duration','genres', 'movie_title','num_voted_users','num_user_for_reviews','title_year','imdb_score']]
data.columns
```

```python
# Define C : mean of imdb score for movies

C=data['imdb_score'].mean()

# Define m: for a movie to be considered in the ranking, it must have garnered more than or least
# 80% of the movies present in the dataset

m=data['num_voted_users'].quantile(0.8)

# Function to compute the IMDB weghted rating for each movie

def weighted_rating(data, m=m ,C=C):
    v=data['num_voted_users']
    R=83668    # R=x['num_voted_users'].mean() 
    return (v/(v+m) * R) + (m /(m+v) *C )
```

```python
def build_chart(data):
    # Ask for preferred genres    
    
    print("Input preferred genre)
    genre=input()
    
    #Ask for lower limit of duration
    
    print("Input shortest duration")
    low_time=int(input())
    
    #Ask for upper time of duration
    
    print("Input longest duration")
    high_time=int(input())
    
    #Ask for lower limit of year
    
    print("Input earliest year")
    low_year=int(input())
    
    #Ask for upper limit of year
    
    print("Input latest year")
    high_year=int(input())
    
# Define a new movies variable to store the preferred movies

    movies=data.copy()

    
# Filter based on the duration and year                          

    movies=movies[(movies['duration']>=low_time) &
                  (movies['duration']<=high_time) &
                  (movies['title_year']>=low_year) &
                  (movies['title_year']<=high_year)]
    
# Only consider movies that have higher than m votes taht m is: m=data['num_voted_users'].quantile(0.8)
   
    q_movies=movies.copy().loc[movies['num_voted_users']>=m]
    
# Filter based on the genres    
    
    q_movies=movies[[genre in x for x in movies['genres']]]
    
# Calculate score using the IMDB formula

    q_movies['score']=q_movies.apply(weighted_rating, axis=1)

# Sort movies in descending order of their score

    q_movies=q_movies.sort_values('score', ascending=False)
    q_movies=q_movies[['director_name','movie_title','genres','num_voted_users','num_user_for_reviews','imdb_score','score']]
    return q_movies
                                          
    
```
